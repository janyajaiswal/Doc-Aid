```python
import PyPDF2
import re
import json

def extract_key_value_pairs(pdf_path):
    key_value_pairs = {}

    with open(pdf_path, 'rb') as pdf_file:
        pdf_reader = PdfReader(pdf_file)
        page_text = ""

        for page_num in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_num]
            page_text += page.extract_text()

        # Define a regex pattern for identifying headings and content
        pattern = r'([A-Z][A-Za-z\s]+):([\s\S]*?)(?=\n[A-Z][A-Za-z\s]+:|$)'

        # Find all matches of the pattern in the text
        matches = re.findall(pattern, page_text)

        for match in matches:
            key = match[0].strip()
            value = match[1].strip()

            # If the key already exists, append the value to a list
            if key in key_value_pairs:
                if isinstance(key_value_pairs[key], list):
                    key_value_pairs[key].append(value)
                else:
                    key_value_pairs[key] = [key_value_pairs[key], value]
            else:
                key_value_pairs[key] = value

    return key_value_pairs

# Replace 'sample.pdf' with the path to your PDF file
pdf_path = '/content/drive/MyDrive/Discharge summaries.pdf'
key_value_pairs = extract_key_value_pairs(pdf_path)

# Convert the dictionary to JSON
json_data = json.dumps(key_value_pairs, indent=4)

# Print or save the JSON data as needed
print(json_data)
```
